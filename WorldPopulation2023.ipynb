## Описание данных

- Country - название стран и зависимых территорий.
- Population 2023 - численность населения в 2023 году
- YearlyChange - процентное ежегодное изменение численности населения
- NetChange - чистое изменение численности населения
- Density(P/Km²)- плотность населения (население на квадратный километр)
- Land Area(Km²) - площадь суши стран / зависимых территорий.
- Migrants(net) - общее число мигрантов
- Fert.Rate - коэффициент фертильности*
- Med.Age - средний возраст населения
- UrbanPop%- процент городского населения
- WorldShare - доля населения в мире




```
* Ферти́льность — способность половозрелого организма производить жизнеспособное потомство.
```

### Изучение распределения населения по регионам

Создание графика распределения населения по регионам

Создание флажков для выбора стран

````python
country_list = data['country'].unique()
country_checkboxes = [widgets.Checkbox(description=country, value=False) for country in country_list]
country_checkboxes_container = widgets.VBox(country_checkboxes)
display(country_checkboxes_container)
````

````python
output = widgets.Output()
````

````python
def plot_population(change):
    selected_country_list = [checkbox.description for checkbox in country_checkboxes if checkbox.value]
    selected_data = data[data['country'].isin(selected_country_list)]
    
    with output:
        output.clear_output(wait=True)
        plt.figure(figsize=(12, 8))
        sns.barplot(y='country', x='population2023', data=selected_data, orient='h', palette='viridis')
        plt.title('Распределение населения по выбранным странам')
        plt.xlabel('Население 2023')
        plt.ylabel('Страна')
        plt.show()
````

````python
for checkbox in country_checkboxes:
    checkbox.observe(plot_population, 'value')
````

````python
display(output)
````

## Анализ зависимостей

### Анализ влияния плотности населения на другие параметры

**Обоснование:** Данный анализ позволяет исследовать, как плотность населения может влиять на различные показатели, такие как общая численность населения, чистое изменение численности, коэффициент фертильности и средний возраст населения.

````python
plt.figure(figsize=(12, 8))
sns.pairplot(data, x_vars='density', y_vars=['population2023', 'netchange', 'fert.rate', 'medianage'],
             kind='scatter', height=3)
plt.suptitle("Анализ влияния плотности населения", y=1.02)
plt.show()
````

#### Вывод: 
Более высокая плотность населения связана с более высоким коэффициентом фертильности, так как на более населенных территориях люди могут предпочитать иметь больше детей. Также, более плотное население может привести к более молодому среднему возрасту населения, так как молодые люди могут иметь более высокую активность в городах с высокой плотностью населения.

### Исследование влияния мигрантов на население

**Обоснование:** Данный анализ позволяет выявить взаимосвязь между количеством мигрантов и общей численностью населения в различных регионах, учитывая процент городского населения.

````python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=data, x='migrants', y='population2023', hue='urbanpop%', palette='coolwarm')
plt.title("Исследование влияния мигрантов на население")
plt.xlabel("Количество мигрантов")
plt.ylabel("Население 2023")
plt.legend(title="Процент городского населения")
plt.show()
````

#### Вывод: 
Более крупные города с более высоким процентом городского населения более привлекательны для мигрантов, что приводит к увеличению общей численности населения в этих городах. Однако возможно, что в некоторых регионах наличие большого количества мигрантов может вызвать конкуренцию за ресурсы и привести к росту численности в меньших городах.

### Исследование влияния площади суши на плотность населения

**Обоснование:** Этот анализ позволяет определить, как площадь суши влияет на плотность населения, с учетом процента городского населения.

````python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, x='land area', y='density', hue='urbanpop%', palette='coolwarm')
plt.title("Исследование влияния площади суши на плотность населения")
plt.xlabel("Площадь суши (Км²)")
plt.ylabel("Плотность населения (чел./Км²)")
plt.legend(title="Процент городского населения")
plt.show()
````

#### Вывод:
Более большие площади суши обеспечивают более низкую плотность населения, так как на таких территориях может быть больше пространства для размещения людей. Однако города с высоким процентом городского населения могут иметь более высокую плотность даже на больших площадях суши.

### Анализ миграций по странам и их влияния на население

### Влияние миграций на численность населения

**Предположения и выводы:**
- График позволит нам оценить, существует ли видимая зависимость между изменением численности населения и числом мигрантов.
- Мы сможем определить, влияют ли миграции на рост или снижение численности населения в разных странах или регионах.

````python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, x='netchange', y='migrants', palette='Set2')
plt.title("Влияние миграций на численность населения")
plt.xlabel("Изменение численности населения за год")
plt.ylabel("Общее число мигрантов")
plt.show()
````

## Анализ доли населения

### Анализ доли населения каждой страны относительно всего мира

**Предположения и выводы:**
- Мы сможем определить страны с наибольшей долей населения относительно всего мира.
- Это позволит нам увидеть, какое количество населения сосредоточено в относительно небольшом числе стран.

````python
data['population_share'] = data['population2023'] / data['population2023'].sum()
top_countries = data.sort_values(by='population_share', ascending=False).head(10)
plt.figure(figsize=(12, 8))
sns.barplot(data=top_countries, y='country', x='population_share', orient='h', palette='Reds')
plt.title("Анализ доли населения каждой страны относительно всего мира")
plt.xlabel("Доля населения")
plt.ylabel("Страна")
plt.show()
````

### Визуализация доли населения для топ-стран

**Предположения и выводы:**
- Круговая диаграмма поможет нам лучше оценить долю населения каждой топ-страны и их вклад в общее население.

````python
plt.figure(figsize=(10, 10))
plt.pie(top_countries['population_share'], labels=top_countries['country'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title("Визуализация доли населения для топ-стран")
plt.show()
````
